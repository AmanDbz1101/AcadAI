```python
import fitz  # PyMuPDF

# Get all documents with category = "Image"
image_docs = [doc for doc in docs if doc.metadata.get("category") == "Image"]


# Open the PDF and create a copy for annotation
pdf_document = fitz.open(file_path)
output_path = "Gated_Attention_highlighted.pdf"

# Highlight each image location
for doc in image_docs:
    page_num = doc.metadata.get("page_number")
    coordinates = doc.metadata.get("coordinates")
    
    if page_num and coordinates:
        page = pdf_document.load_page(page_num - 1)  # Convert to 0-indexed
        points = coordinates.get("points", [])
        layout_width = coordinates.get("layout_width")
        layout_height = coordinates.get("layout_height")
        
        if points and len(points) >= 2 and layout_width and layout_height:
            # Get page dimensions
            page_rect = page.rect
            page_width = page_rect.width
            page_height = page_rect.height
            
            # Calculate scaling factors
            scale_x = page_width / layout_width
            scale_y = page_height / layout_height
            
            # Transform coordinates from PixelSpace to PDF space
            x_coords = [p[0] * scale_x for p in points]
            y_coords = [p[1] * scale_y for p in points]
            
            rect = fitz.Rect(min(x_coords), min(y_coords), max(x_coords), max(y_coords))
            
            # Highlight the rectangle
            page.draw_rect(rect, color=(1, 1, 0), width=2)  # Yellow border

# Save the annotated PDF
pdf_document.save(output_path)
pdf_document.close()

print(f"PDF with highlighted images saved to {output_path}")
```

## Formula

```python
from IPython.display import HTML, display

# Get all documents with category = "Formula"
formula_docs = [doc for doc in docs if doc.metadata.get("category") == "Formula"]

# Display formulas with highlighting
for doc in formula_docs:
    print(f"Page {doc.metadata.get('page_number')}: Formula")
    if doc.metadata.get("text_as_html"):
        display(HTML(doc.metadata.get("text_as_html")))
    else:
        print(doc.page_content)
    print("-" * 50)

# Open the PDF and highlight formula locations
pdf_document = fitz.open(file_path)
output_path_formula = "Gated_Attention_formulas.pdf"

for doc in formula_docs:
    page_num = doc.metadata.get("page_number")
    coordinates = doc.metadata.get("coordinates")
    
    if page_num and coordinates:
        page = pdf_document.load_page(page_num - 1)
        points = coordinates.get("points", [])
        layout_width = coordinates.get("layout_width")
        layout_height = coordinates.get("layout_height")
        
        if points and len(points) >= 2 and layout_width and layout_height:
            page_rect = page.rect
            page_width = page_rect.width
            page_height = page_rect.height
            
            scale_x = page_width / layout_width
            scale_y = page_height / layout_height
            
            x_coords = [p[0] * scale_x for p in points]
            y_coords = [p[1] * scale_y for p in points]
            
            rect = fitz.Rect(min(x_coords), min(y_coords), max(x_coords), max(y_coords))
            page.draw_rect(rect, color=(1, 0, 0), width=2)  # Red border for formulas

pdf_document.save(output_path_formula)
pdf_document.close()

print(f"PDF with highlighted formulas saved to {output_path_formula}")
```

## Saving figures in the pdf

```python
# Crop all images using their coordinates and save to ../output/figures


pdf_document = fitz.open(file_path)
for idx, doc in enumerate(image_docs, start=1):
    coords = doc.metadata.get("coordinates")
    page_num = doc.metadata.get("page_number")
    if not coords or not page_num:
        continue

    points = coords.get("points", [])
    layout_width = coords.get("layout_width")
    layout_height = coords.get("layout_height")
    if not (points and layout_width and layout_height):
        continue

    page = pdf_document.load_page(page_num - 1)
    page_rect = page.rect
    scale_x = page_rect.width / layout_width
    scale_y = page_rect.height / layout_height

    x_coords = [p[0] * scale_x for p in points]
    y_coords = [p[1] * scale_y for p in points]
    rect = fitz.Rect(min(x_coords), min(y_coords), max(x_coords), max(y_coords))

    pix = page.get_pixmap(clip=rect)
    out_path = f"../output/figures/image_page{page_num}_{idx}.png"
    pix.save(out_path)
    print(f"Saved: {out_path}")

pdf_document.close()
```

```python
import base64
import io

import fitz
from PIL import Image

from IPython.display import Image as IPImage
from IPython.display import display
```

```python
pdf_document = fitz.open(file_path)
page = pdf_document.load_page(2 - 1)  # input is one-indexed
pix = page.get_pixmap()
img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

buffer = io.BytesIO()
img.save(buffer, format="PNG")

```

```python
image = base64.b64encode(buffer.getvalue()).decode("utf-8")
```

```python
display(IPImage(data=base64.b64decode(image)))
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
from langchain_core.messages import HumanMessage

query = "What are the practical difficulties for taking advantages of recent advances in DL-based methods?"

message = HumanMessage(
    content=[
        {"type": "text", "text": query},
        {
            "type": "image_url",
            "image_url": {"url": f"data:image/jpeg;base64,{image}"},
        },
    ],
)
response = llm.invoke([message])
print(response.content)
```
