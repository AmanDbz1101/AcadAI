```python
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.datamodel.base_models import InputFormat
import json

def extract_research_paper_with_bboxes(pdf_path, output_json_path):
    """
    Extract all elements from a research paper with bounding boxes.
    
    Args:
        pdf_path: Path to the PDF file
        output_json_path: Path to save the extraction results
    
    Returns:
        Dictionary with extracted elements and their bounding boxes
    """
    
    # Configure for digital PDFs
    pipeline_options = PdfPipelineOptions(
        do_ocr=False,                 # Not needed for digital PDFs
        do_table_structure=True,      # Extract tables
        do_formula_enrichment=False,   # Detect and enrich formulas (corrected option name)
        do_code_enrichment=False,     # Usually not needed
    )
    
    # Create converter with proper format_options
    converter = DocumentConverter(
        format_options={
            InputFormat.PDF: PdfFormatOption(
                pipeline_options=pipeline_options
            )
        }
    )
    
    # Convert document
    result = converter.convert(pdf_path)
    doc = result.document
    
    # Storage for all extracted elements
    extracted_data = {
        "title": None,
        "authors": [],
        "abstract": None,
        "sections": [],
        "tables": [],
        "figures": [],
        "references": [],
        "all_text_elements": []
    }
    
    # Extract text elements with bounding boxes
    for text_item in doc.texts:
        element_data = {
            "type": text_item.label,
            "text": text_item.text,
            "self_ref": text_item.self_ref,
            "locations": []
        }
        
        # Get all provenance information
        if text_item.prov:
            for prov in text_item.prov:
                location = {
                    "page": prov.page_no,
                    "bbox": {
                        "left": prov.bbox.l,
                        "top": prov.bbox.t,
                        "right": prov.bbox.r,
                        "bottom": prov.bbox.b,
                        "coord_origin": str(prov.bbox.coord_origin)
                    }
                }
                element_data["locations"].append(location)
        
        # Categorize based on label (case-insensitive)
        label = text_item.label.lower()
        
        if "title" in label and not extracted_data["title"]:
            extracted_data["title"] = element_data
        elif "abstract" in label:
            extracted_data["abstract"] = element_data
        elif "section" in label or "heading" in label:
            extracted_data["sections"].append(element_data)
        elif "reference" in label or "citation" in label:
            extracted_data["references"].append(element_data)
        
        # Add to comprehensive list
        extracted_data["all_text_elements"].append(element_data)
    
    # Extract tables with bounding boxes
    for table_item in doc.tables:
        caption_str = table_item.caption_text(doc)
        
        table_data = {
            "type": "table",
            "caption": caption_str if caption_str else None,
            "self_ref": table_item.self_ref,
            "locations": [],
            "data": None
        }
        
        if table_item.prov:
            for prov in table_item.prov:
                location = {
                    "page": prov.page_no,
                    "bbox": {
                        "left": prov.bbox.l,
                        "top": prov.bbox.t,
                        "right": prov.bbox.r,
                        "bottom": prov.bbox.b,
                        "coord_origin": str(prov.bbox.coord_origin)
                    }
                }
                table_data["locations"].append(location)
        
        # Extract table data as list of dicts (rows)
        if table_item.data:
            try:
                table_data["data"] = table_item.export_to_dataframe().to_dict('records')
            except Exception as e:
                table_data["data"] = {"error": f"Failed to export table: {str(e)}"}
        
        extracted_data["tables"].append(table_data)
    
    # Extract figures with bounding boxes
    for picture_item in doc.pictures:
        caption_str = picture_item.caption_text(doc)
        
        figure_data = {
            "type": "figure",
            "caption": caption_str if caption_str else None,
            "self_ref": picture_item.self_ref,
            "locations": []
        }
        
        if picture_item.prov:
            for prov in picture_item.prov:
                location = {
                    "page": prov.page_no,
                    "bbox": {
                        "left": prov.bbox.l,
                        "top": prov.bbox.t,
                        "right": prov.bbox.r,
                        "bottom": prov.bbox.b,
                        "coord_origin": str(prov.bbox.coord_origin)
                    }
                }
                figure_data["locations"].append(location)
        
        extracted_data["figures"].append(figure_data)
    
    # Save to JSON
    with open(output_json_path, 'w', encoding='utf-8') as f:
        json.dump(extracted_data, f, indent=2, ensure_ascii=False)
    
    return extracted_data


# Example usage
if __name__ == "__main__":
    data = extract_research_paper_with_bboxes(
        pdf_path="MemGPT.pdf",
        output_json_path="extracted_data.json"
    )
    
    print(f"Extracted {len(data['all_text_elements'])} text elements")
    print(f"Found {len(data['tables'])} tables")
    print(f"Found {len(data['figures'])} figures")
    print(f"Found {len(data['sections'])} section headings")
```

## Highlighting Images and Formulas with Docling BBoxes

```python
import fitz  # PyMuPDF

def docling_bbox_to_pymupdf(bbox, page_height):
    """
    Convert Docling BOTTOMLEFT coordinates to PyMuPDF TOPLEFT.
    
    Args:
        bbox: Docling bbox object with l, t, r, b properties
        page_height: PDF page height (page.rect.height)
    
    Returns:
        fitz.Rect in PyMuPDF coordinate system
    """
    # Convert y-coordinates: flip vertically
    top_pymupdf = page_height - bbox.t
    bottom_pymupdf = page_height - bbox.b
    
    return fitz.Rect(bbox.l, top_pymupdf, bbox.r, bottom_pymupdf)
```

```python
# Highlight images and formulas in the PDF
pdf_path = "MemGPT.pdf"
output_path = "MemGPT_highlighted_docling.pdf"

# Open the PDF
pdf_document = fitz.open(pdf_path)

# Track counts
image_count = 0
formula_count = 0

# Highlight all pictures/figures (Yellow)
for picture_item in doc.pictures:
    if picture_item.prov:
        for prov in picture_item.prov:
            page_num = prov.page_no
            bbox = prov.bbox
            
            # Load the page (convert from 1-indexed to 0-indexed)
            page = pdf_document.load_page(page_num - 1)
            
            # Convert Docling bbox to PyMuPDF rect
            rect = docling_bbox_to_pymupdf(bbox, page.rect.height)
            
            # Highlight with yellow border
            page.draw_rect(rect, color=(1, 1, 0), width=2)
            image_count += 1

# Highlight all formulas (Red)
for text_item in doc.texts:
    label = text_item.label.lower()
    
    # Check if it's a formula/equation
    if "formula" in label or "equation" in label:
        if text_item.prov:
            for prov in text_item.prov:
                page_num = prov.page_no
                bbox = prov.bbox
                
                # Load the page
                page = pdf_document.load_page(page_num - 1)
                
                # Convert Docling bbox to PyMuPDF rect
                rect = docling_bbox_to_pymupdf(bbox, page.rect.height)
                
                # Highlight with red border
                page.draw_rect(rect, color=(1, 0, 0), width=2)
                formula_count += 1

# Save the annotated PDF
pdf_document.save(output_path)
pdf_document.close()

print(f"✅ PDF with highlighted elements saved to {output_path}")
print(f"📊 Highlighted {image_count} images (yellow)")
print(f"🔢 Highlighted {formula_count} formulas (red)")
```

## Cropping Images and Formulas

```python
import os

# Create output directories if they don't exist
os.makedirs("../output/figures", exist_ok=True)
os.makedirs("../output/formulas", exist_ok=True)

# Crop and save all images
pdf_document = fitz.open(pdf_path)

print("Cropping images...")
for idx, picture_item in enumerate(doc.pictures, start=1):
    if picture_item.prov:
        for prov in picture_item.prov:
            page_num = prov.page_no
            bbox = prov.bbox
            
            # Load the page
            page = pdf_document.load_page(page_num - 1)
            
            # Convert Docling bbox to PyMuPDF rect
            rect = docling_bbox_to_pymupdf(bbox, page.rect.height)
            
            # Crop the image
            pix = page.get_pixmap(clip=rect)
            
            # Save with page number and index
            output_filename = f"../output/figures/image_page{page_num}_{idx}.png"
            pix.save(output_filename)
            print(f"  ✓ Saved: {output_filename}")

print("\nCropping formulas...")
formula_idx = 0
for text_item in doc.texts:
    label = text_item.label.lower()
    
    # Check if it's a formula/equation
    if "formula" in label or "equation" in label:
        if text_item.prov:
            for prov in text_item.prov:
                formula_idx += 1
                page_num = prov.page_no
                bbox = prov.bbox
                
                # Load the page
                page = pdf_document.load_page(page_num - 1)
                
                # Convert Docling bbox to PyMuPDF rect
                rect = docling_bbox_to_pymupdf(bbox, page.rect.height)
                
                # Crop the formula
                pix = page.get_pixmap(clip=rect)
                
                # Save with page number and index
                output_filename = f"../output/formulas/formula_page{page_num}_{formula_idx}.png"
                pix.save(output_filename)
                print(f"  ✓ Saved: {output_filename}")

pdf_document.close()
print("\n✅ All images and formulas cropped and saved!")
```

## Comparison: Docling vs UnstructuredLoader

Key advantages of using Docling's bounding boxes:
- ✅ **No scaling calculations needed** - coordinates are already in PDF points
- ✅ **Simpler bbox structure** - direct `l/t/r/b` properties
- ✅ **Clean coordinate conversion** - only need to flip Y-axis for PyMuPDF
- ✅ **Consistent coordinate system** - has `coord_origin` property for clarity

Only requirement: Convert from BOTTOMLEFT to TOPLEFT coordinates using `page_height - y`
