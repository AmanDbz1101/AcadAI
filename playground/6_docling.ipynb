```python
# Install and import required libraries for Docling PDF parsing
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.datamodel.accelerator_options import AcceleratorOptions, AcceleratorDevice
import os

# Path to the PDF file
pdf_path = os.path.abspath("MemGPT.pdf")

# Configure pipeline options for maximum metadata extraction
pipeline_options = PdfPipelineOptions()
pipeline_options.do_ocr = False
pipeline_options.do_table_structure = True
pipeline_options.do_code_enrichment = True
pipeline_options.do_formula_enrichment = True
pipeline_options.generate_page_images = False
pipeline_options.generate_picture_images = False
pipeline_options.images_scale = 1.0

# Force GPU usage (CUDA for NVIDIA GPUs)
pipeline_options.accelerator_options = AcceleratorOptions(
    num_threads=8,
    device=AcceleratorDevice.CPU
)

# Set up the DocumentConverter with custom options
converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
    }
)
```

```python
# Parse the PDF
result = converter.convert(pdf_path)
doc = result.document
```

```python
print(doc.export_to_markdown())
```

## Simple markdown output

```python
from docling.document_converter import DocumentConverter

source = "MemGPT.pdf"
converter = DocumentConverter()
doc = converter.convert(source).document
```

```python
print(doc.export_to_markdown())

```

```python
with open('MemGPT.md', 'w', encoding='utf-8') as f:
    f.write(doc.export_to_markdown())
```

```python
# Extract document structure information
# Get all headers/sections
headers = []
for item, level in doc.iterate_items():
    if item.label == "section_header":
        headers.append({
            "text": item.text,
            "level": level
        })

# Count all element types
captions = []
codes = []
footnotes = []
list_items = []
pictures = []
tables = []
texts = []
formulas = []

for item, level in doc.iterate_items():
    if item.label == "caption":
        captions.append(item)
    elif item.label == "code":
        codes.append(item)
    elif item.label == "footnote":
        footnotes.append(item)
    elif item.label == "list_item":
        list_items.append(item)
    elif item.label == "picture":
        pictures.append(item)
    elif item.label == "table":
        tables.append(item)
    elif item.label == "text":
        texts.append(item)
    elif item.label in ["formula", "equation"]:
        formulas.append(item)

# Display results
print("=" * 60)
print(f"DOCUMENT STRUCTURE ANALYSIS")
print("=" * 60)
print(f"\n📑 SECTIONS/HEADERS: {len(headers)}")
print("-" * 60)
for i, header in enumerate(headers, 1):
    indent = "  " * (header['level'] - 1)
    print(f"{indent}{i}. {header['text']}")

print(f"\n📊 ELEMENT STATISTICS:")
print("-" * 60)
print(f"📝 Text blocks: {len(texts)}")
print(f"📋 Tables: {len(tables)}")
print(f"🖼️  Pictures: {len(pictures)}")
print(f"📑 List items: {len(list_items)}")
print(f"💬 Captions: {len(captions)}")
print(f"💻 Code blocks: {len(codes)}")
print(f"🔢 Footnotes: {len(footnotes)}")
print(f"🧮 Formulas: {len(formulas)}")
print("=" * 60)
```

```python
# Print all code blocks found in the document
print("=" * 60)
print(f"CODE BLOCKS FOUND: {len(codes)}")
print("=" * 60)

for i, code_item in enumerate(codes, 1):
    print(f"\n{'='*60}")
    print(f"CODE BLOCK #{i}")
    print(f"{'='*60}")
    print(code_item.text)
    print(f"{'='*60}\n")
```
